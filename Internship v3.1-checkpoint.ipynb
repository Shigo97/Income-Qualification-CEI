```python
# Function to automatically get the list of Int/Float Column(s) containing Null value: 
def get_int_float_dtype_null_column_list(df_data):
    # Getting the list of column(s) with datatype as int64 or float64 and having null value
    return df_data.columns[((df_data.dtypes == "int64") | (df_data.dtypes == "float64")) & (df_data.isna().any())].tolist()
```

```python
# Function to automatically get Object Dtype categorical column(s) (i.e, columns having <= (say 10) diff type of data):
def get_category_column_list(df_data, object_column_list):
    # Though in this case 'ocean_proximity' is the only column falling in this list having 5 distinct values. 
    #  I still have set the CATEGORY_COLUMN_MAX_DISTINCT_VAL as 10 for generic function purpose
    CATEGORY_COLUMN_MAX_DISTINCT_VAL = 10
    category_column_list = []
    category_column_val_list = []
    for col in object_column_list:
        col_val_list_count = df_data[col].value_counts()
        if col_val_list_count.size <= CATEGORY_COLUMN_MAX_DISTINCT_VAL:
            category_column_list.append(col)
            category_column_val_list.append(df_data[col].value_counts().index.to_list())
            print("\nDistinct Values of the Category [", col, "]:\n", df_data[col].value_counts())
            print(df_data[col].value_counts().index.to_list())
    return category_column_list, category_column_val_list
```

```python
# This fn creates additional "columnValue" column(s) for the corresponding object categorical column(s) 
#      and corresponding data value(s) for all except the last data value ():
def apply_category_column_encoding(df_data, category_column_list, category_column_val_list):
    category_column_idx = -1   
    for col in category_column_list:
        category_column_idx = category_column_idx + 1   
        # One Hot Encoding will be done all values minus 1 
        count_category_column_val_list = len(category_column_val_list[category_column_idx]) -1
        for idx in range(count_category_column_val_list):
            df_data[category_column_val_list[category_column_idx][idx]] = df_data[col].apply((lambda x: 1 if x == category_column_val_list[category_column_idx][idx] else 0))
# Note: Parameter <category_column_list> is ['ocean_proximity'] & category_column_val_list is [['<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'NEAR BAY', 'ISLAND']]
apply_category_column_encoding(housing_df, category_column_list, category_column_val_list)
```

```python
# The variables on the first line are of int datatype and shall be used in order to tell how much of a sample we want. 
# While between lines 2 and 4 are of DataFrame datatype each is a slice of the DataFrame containing only one type of class. 
# Lastly, on lines 5 and 6, we will re-assign the DataFrames to new ones but we will apply the sample function to it and pass to it the int value of the least class, in this case, class_1.

Class_4, Class_3, Class_2, Class_1 = income_train_df['Target'].value_counts()
c4 = income_train_df[income_train_df['Target'] == 4]
c3 = income_train_df[income_train_df['Target'] == 3]
c2 = income_train_df[income_train_df['Target'] == 2]
c1 = income_train_df[income_train_df['Target'] == 1]
```

```python
train_1 = c4.sample(Class_1)
train_2 = c3.sample(Class_1)
train_3 = c2.sample(Class_1)
```

```python
# last we will concatenate the various dataframes with the original dataframe i.e. the df with the minority class

undersampled_df = pd.concat([train_1, train_2, train_3, c1], axis = 0)
```

```python
# Oversampling 

Class_4, Class_3, Class_2, Class_1 = income_train_df['Target'].value_counts()
c4 = income_train_df[income_train_df['Target'] == 4]
c3 = income_train_df[income_train_df['Target'] == 3]
c2 = income_train_df[income_train_df['Target'] == 2]
c1 = income_train_df[income_train_df['Target'] == 1]

# Now sample and fit with reference to the majority class

train_1 = c1.sample(Class_4, replace = True)
train_2 = c2.sample(Class_4, replace = True)
train_3 = c3.sample(Class_4, replace = True)

# next step is to conatenate the various dfs to the original df i.e. the df with the majority class and create a new df

oversampled_df = pd.concat([train_1, train_2, train_3, c4], axis = 0)
print(oversampled_df.Target.value_counts())
```
